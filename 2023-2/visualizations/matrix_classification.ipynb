```python
import plotly.graph_objects as go
import numpy as np
```

```python
quadratic_form = lambda x, M: np.dot(x, np.dot(M, x))
```

```python
def get_surf_points(M,lb,ub,n):
    x_values = np.linspace(lb, ub, n)
    y_values = np.linspace(lb, ub, n)
    X, Y = np.meshgrid(x_values, y_values)
    Z = np.zeros_like(X)
    for i in range(len(x_values)):
        for j in range(len(y_values)):
            x = np.array([x_values[i], y_values[j]])
            Z[i, j] = quadratic_form(x, M)
    
    return X,Y,Z
```

```python
def plot_quadratic(M,lb=-5,ub=5,n=100):
    X,Y,Z = get_surf_points(M,lb,ub,n)
    # Create an interactive surface plot
    fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
    fig.update_layout(width=1000, height=1000)
    fig.update_layout(scene=dict(zaxis_title='Quadratic Form'))
    fig.update_layout(scene=dict(aspectmode="cube"))

    # Show the plot in an interactive viewer
    fig.show()
```

```python
M1 = [[-1,0],
      [0,0]]

plot_quadratic(M1)
```
