```python
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
```

```python
quadratic_form = lambda x, M: np.dot(x, np.dot(M, x))
```

```python
f = lambda x: 8 + np.dot(np.array([11,2]),[x[0]-2,x[1]-1]) + 1/2*np.dot(np.dot(np.array([x[0]-2,x[1]-1]),np.array([[12,-1],[-1,4]])),np.array([x[0]-2,x[1]-1]))
```

```python
f_o = lambda x: x[0]**3+2*x[1]**2-x[0]*x[1]
```

```python
def get_surf_points2(f,lb,ub,n):
    x_values = np.linspace(lb, ub, n)
    y_values = np.linspace(lb, ub, n)
    X, Y = np.meshgrid(x_values, y_values)
    Z = np.zeros_like(X)
    for i in range(len(x_values)):
        for j in range(len(y_values)):
            x = np.array([x_values[i], y_values[j]])
            Z[i, j] = f(x)
    
    return X,Y,Z
```

```python
def plot_f(f1,f2,lb=-5,ub=5,n=100):
    X1,Y1,Z1 = get_surf_points2(f1,lb,ub,n)
    X2,Y2,Z2 = get_surf_points2(f2,lb,ub,n)
    # Create an interactive surface plot
    fig = go.Figure(data=[go.Surface(z=Z1, x=X1, y=Y1),go.Surface(z=Z2, x=X2, y=Y2)])
    
    fig.update_layout(width=1000, height=1000)
    fig.update_layout(scene=dict(zaxis_title='Quadratic Form'))
    fig.update_layout(scene=dict(aspectmode="cube"))

    fig.show()
```

```python
plot_f(f,f_o,0.9,2.1,100)
plt.show()
```

```python
def get_surf_points(M,lb,ub,n):
    x_values = np.linspace(lb, ub, n)
    y_values = np.linspace(lb, ub, n)
    X, Y = np.meshgrid(x_values, y_values)
    Z = np.zeros_like(X)
    for i in range(len(x_values)):
        for j in range(len(y_values)):
            x = np.array([x_values[i], y_values[j]])
            Z[i, j] = quadratic_form(x, M)
    
    return X,Y,Z
```

```python
def plot_quadratic(M,lb=-5,ub=5,n=100):
    X,Y,Z = get_surf_points(M,lb,ub,n)
    # Create an interactive surface plot
    fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
    fig.update_layout(width=1000, height=1000)
    fig.update_layout(scene=dict(zaxis_title='Quadratic Form'))
    fig.update_layout(scene=dict(aspectmode="cube"))

    # Show the plot in an interactive viewer
    fig.show()
```

```python
M1 = [[-1,0],
      [0,0]]

plot_quadratic(M1)
```
